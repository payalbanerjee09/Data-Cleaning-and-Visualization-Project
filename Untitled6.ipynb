```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- STEP 1: GENERATING REALISTIC HUMAN TRANSACTION DATA ---
np.random.seed(29)

indian_names = [
    "Payal Banerjee", "abhiiee", "Amit Verma", "Sneha Iyer",
    "Vikram Malhotra", "Pooja Patel", "Rohan Das", "Ananya Reddy",
    "Arjun Kapoor", "Meera Nair", "Siddharth Rao", "Kriti Joshi"
]

cities = ["Kolkata", "Mumbai", "Delhi", "Bangalore", "Hyderabad", "Chennai"]
products = ["Wireless Mouse", "Mechanical Keyboard", "1080p Monitor", "USB-C Hub", "Ergonomic Chair", "Noise Cancelling Headphones"]
categories = ["Electronics", "Electronics", "Hardware", "Accessories", "Furniture", "Audio"]

prod_to_cat = dict(zip(products, categories))
prod_prices = {
    "Wireless Mouse": 1200,
    "Mechanical Keyboard": 4500,
    "1080p Monitor": 11500,
    "USB-C Hub": 1800,
    "Ergonomic Chair": 15000,
    "Noise Cancelling Headphones": 8500
}

n_rows = 350

data = {
    "TransactionID": [f"TXN2026{i:03d}" for i in range(1, n_rows + 1)],
    "CustomerName": np.random.choice(indian_names, size=n_rows),
    "City": np.random.choice(cities, size=n_rows),
    "Product": np.random.choice(products, size=n_rows),
    "Quantity": np.random.choice([1, 2, 3, 4], p=[0.5, 0.3, 0.15, 0.05], size=n_rows),
    "Discount_Pct": np.random.choice([0, 5, 10, 15, 20], p=[0.4, 0.3, 0.15, 0.1, 0.05], size=n_rows),
    "PaymentMode": np.random.choice(["UPI", "Credit Card", "Debit Card", "Net Banking"], p=[0.6, 0.2, 0.1, 0.1], size=n_rows)
}

df = pd.DataFrame(data)

df["Category"] = df["Product"].map(prod_to_cat)
df["BasePrice"] = df["Product"].map(prod_prices)
df["GrossAmount"] = df["BasePrice"] * df["Quantity"]
df["DiscountAmount"] = (df["GrossAmount"] * df["Discount_Pct"]) / 100
df["NetAmount"] = df["GrossAmount"] - df["DiscountAmount"]

nan_indices = np.random.choice(df.index, size=12, replace=False)
df.loc[nan_indices, "Discount_Pct"] = np.nan

print(">>> Initial Dataset Audit:")
print(df.head(10))
print("\n" + "="*50 + "\n")

# --- STEP 2: DATA PREPROCESSING & CLEANING ---
print(">>> Checking for Structural Anomalies / Missing Values:")
print(df.isnull().sum())

most_common_discount = df["Discount_Pct"].mode()[0]
df["Discount_Pct"].fillna(most_common_discount, inplace=True)

df["DiscountAmount"] = (df["GrossAmount"] * df["Discount_Pct"]) / 100
df["NetAmount"] = df["GrossAmount"] - df["DiscountAmount"]

print("\n>>> Sanity check after imputation (Missing fields remaining):")
print(df.isnull().sum())
print("\n" + "="*50 + "\n")

# --- STEP 3: EXPLORATORY DATA ANALYSIS & CORE AGGREGATIONS ---
print(">>> Summary Statistics for Financial Outliers:")
print(df[["Quantity", "GrossAmount", "NetAmount"]].describe())

print("\n>>> Top Spenders (Human Customer Analysis):")
top_spenders = df.groupby("CustomerName")["NetAmount"].sum().reset_index()
top_spenders = top_spenders.sort_values(by="NetAmount", ascending=False).reset_index(drop=True)
print(top_spenders.head(5))

print("\n>>> Regional Performance Matrices (City vs Net Revenue):")
regional_revenue = df.groupby("City")["NetAmount"].sum().sort_values(ascending=False).reset_index()
print(regional_revenue)

print("\n>>> Category-wise Product Movements:")
category_metrics = df.groupby("Category").agg(
    TotalUnitsSold=('Quantity', 'sum'),
    TotalNetRevenue=('NetAmount', 'sum'),
    AverageDiscount=('Discount_Pct', 'mean')
).reset_index()
print(category_metrics)
print("\n" + "="*50 + "\n")

# --- STEP 4: DATA VISUALIZATION ENGINE ---
sns.set_style("ticks")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

sns.histplot(df["NetAmount"], kde=True, color="#2b5c8f", ax=axes[0, 0], bins=20)
axes[0, 0].set_title("Distribution of Transaction Net Values", fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel("Net Amount (INR)")

sns.barplot(data=regional_revenue, x="City", y="NetAmount", palette="mako", ax=axes[0, 1])
axes[0, 1].set_title("Total Revenue Generation by City", fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel("Accumulated Revenue (INR)")
axes[0, 1].tick_params(axis='x', rotation=15)

payment_counts = df["PaymentMode"].value_counts().reset_index()
axes[1, 0].pie(payment_counts["count"], labels=payment_counts["PaymentMode"], autopct='%1.1f%%',
               colors=["#4f81bd", "#c0504d", "#9bbb59", "#8064a2"], startangle=140)
axes[1, 0].set_title("Preferred Customer Payment Gateways", fontsize=12, fontweight='bold')

sns.boxplot(data=df, x="Category", y="NetAmount", palette="Set2", ax=axes[1, 1])
axes[1, 1].set_title("Transaction Volume and Range by Category", fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel("Net Amount Per Order (INR)")

plt.tight_layout()
plt.show()

print("\n>>> All data verification pipelines executed successfully. Visual patterns mapped.")
```
