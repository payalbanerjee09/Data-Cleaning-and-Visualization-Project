```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# CODE PART 1: DATASET GENERATION
# ==========================================
np.random.seed(42)
data = {
    'CustomerID': range(101, 201),
    'Age': np.random.randint(18, 70, size=100),
    'Annual_Income_k$': np.random.randint(15, 130, size=100),
    'Spending_Score': np.random.randint(1, 100, size=100),
    'Gender': np.random.choice(['Male', 'Female'], size=100),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Books'], size=100)
}
df = pd.DataFrame(data)

print("=== PART 1: DATASET PEHLE 5 ROWS ===")
print(df.head())
print("\n" + "="*40 + "\n")

# ==========================================
# CODE PART 2: STATISTICAL SUMMARY
# ==========================================
print("=== PART 2: DATA INFO ===")
print(df.info())

print("\n=== PART 2: MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== PART 2: STATISTICAL SUMMARY ===")
print(df.describe())
print("\n" + "="*40 + "\n")

# ==========================================
# CODE PART 3: VISUALIZATIONS & CHARTS
# ==========================================
print("=== PART 3: GENERATING GRAPHS (CHARTS) ===")

# Graph 1: Age & Income Distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, color='blue')
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['Annual_Income_k$'], kde=True, color='green')
plt.title('Annual Income Distribution')
plt.tight_layout()
plt.show()

# Graph 2: Category wise Spending Score
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Spending_Score', data=df, hue='Gender', palette='Set2')
plt.title('Spending Score by Category and Gender')
plt.show()

# Graph 3: Correlation Matrix Heatmap
plt.figure(figsize=(6, 4))
numeric_df = df[['Age', 'Annual_Income_k$', 'Spending_Score']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (Key Influencing Factors)')
plt.show()
```
