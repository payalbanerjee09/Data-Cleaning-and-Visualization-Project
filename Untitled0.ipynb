```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Online raw sales dataset load kar rahe hain
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
df = pd.read_csv(url, sep='\t')

# Dekhte hain data dikhta kaisa hai
df.head()
```

```python
# 1. Check karte hain kahin koi data khali (missing) toh nahi hai
print("Missing values in each column:\n", df.isnull().sum())

# 2. 'choice_description' mein khali data ko 'Regular' se bhar dete hain
df['choice_description'] = df['choice_description'].fillna('Regular')

# 3. 'item_price' se '$' hata kar use float (number) mein convert karte hain
df['item_price'] = df['item_price'].str.replace('$', '').astype(float)

# 4. Check karte hain koi duplicate row toh nahi hai, hai toh hata dete hain
df.drop_duplicates(inplace=True)

print("\n--- Data Clean Ho Gaya! ---")
df.head()
```

```python
# Ek bada aur saaf graph set karte hain
plt.figure(figsize=(10, 6))

# Top 5 sabse zyada bikne waale items ka graph
top_items = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)

sns.barplot(x=top_items.values, y=top_items.index, palette='Greens_r')
plt.title('Top 5 Most Ordered Items 📊', fontsize=16)
plt.xlabel('Total Quantity Ordered')
plt.ylabel('Item Name')
plt.show()
```
